```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

insurance = pd.read_csv('./insurance.csv')
insurance.head()
```

Content:
Columns

* age: age of primary beneficiary

* sex: insurance contractor gender, female, male

* bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

* children: Number of children covered by health insurance / Number of dependents

* smoker: Smoking

* region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

* charges: Individual medical costs billed by health insurance

am trying to figure out how each one of these effect the health insurance charges 

```python
x = insurance['age'].values
y = insurance['charges'].values 
x=x.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=100)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
plt.scatter(x_train,y_train,color='blue')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('The effect of age on insurance charges')
plt.show()
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
y_test
```

```python
y_predict
```

```python
new_person_insurance_charges = np.array([30]).reshape(-1,1)
new_person_insurance_charges
```

```python
model.predict(new_person_insurance_charges)
```

```python
new_person_insurance_charges = np.array([25]).reshape(-1,1)
model.predict(new_person_insurance_charges)
```

```python
model.score(x_train,y_train)*100
```

```python
plt.scatter(x_train,y_train,color='blue')
plt.plot(x_test,y_predict,color='red')
plt.xlabel('Age')
plt.ylabel('Insurance charges')
plt.title('The effect of age on insurance charges')
plt.show()
```
